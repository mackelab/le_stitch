```python
%matplotlib inline
import numpy as np
from scipy.io import loadmat
import glob, os

absolute_code_path = '/home/marcel/Desktop/Projects/Stitching/code/code_le_stitch/iterSSID/python/'
os.chdir(absolute_code_path +'core')
import stitching_ssid

from scipy.io import savemat # store results for comparison with Matlab code   
```

# Generate SSID initialisations for illustration #1

```python
relative_data_path = '../../../../../results/cosyne_poster/illustration_1/data'
os.chdir(relative_data_path)
filenames = glob.glob("*.npz")
num_exps = len(filenames)
idx_exps = range(num_exps)


pi_method = 'heuristic'
num_iter = 300

for i in idx_exps:
    
    file = filenames[i]
    
    loadfile = np.load(file)

    data = loadfile['y']
    T,p = data.shape
    stateseq = loadfile['x']
    n   = stateseq.shape[1]
    obs_scheme = {'sub_pops' : loadfile['sub_pops'], 
                  'obs_pops' : loadfile['obs_pops'], 
                  'obs_time' : loadfile['obs_time']}

    seq, seq_f = obs_scheme.copy(), {'sub_pops' : (np.arange(p),), 'obs_pops' : (0,), 'obs_time' : (T,)} 
    seq_f['y'], seq['y'] = data, data.copy()

    params, SIGfp, SIGyy, lPSIGfp, lPSIGyy, SIGfp_f, SIGyy_f = \
        stitching_ssid.run_exp_iterSSID(seq, seq_f, n, num_iter=num_iter,pi_method=pi_method, plot_flag=True)
        
    params_naive = stitching_ssid.stabilize_A(params[0])
    params_final = stitching_ssid.stabilize_A(params[-1])
        
    save_file = '../init/init_'+file 
    save_file_m = {'params': params_final,
                   'params_naive': params_naive,
                   'params_hist' : params,
                   'SIGfp' : SIGfp,
                   'SIGyy' : SIGyy,
                   'lPSIGfp' : lPSIGfp,
                   'lPSIGyy' : lPSIGyy,
                   'data': seq['y'],                   
                   'p' : p,
                   'n' : n,
                   'T' : T,
                   'filename' : file,
                   'num_iter' : num_iter,
                   'pi_method' : pi_method
                   }

    savemat(save_file,save_file_m) # does the actual saving
    
    np.savez(save_file, params=params_final,
                        params_naive=params_naive,
                        params_hist=params, 
                        SIGfp = SIGfp,
                        SIGyy = SIGyy,
                        lPSIGfp = lPSIGfp,
                        lPSIGyy = lPSIGyy,
                        data = seq['y'],                   
                        p = p,
                        n = n,
                        T = T,
                        filename = file,
                        num_iter = num_iter,
                        pi_method = pi_method,
                        sub_pops=obs_scheme['sub_pops'],            
                        obs_time=obs_scheme['obs_time'],            
                        obs_pops=obs_scheme['obs_pops'])               
  
```

```python
relative_data_path = '../../../../../results/cosyne_poster/simulation_2/data'
os.chdir(relative_data_path)
filenames = glob.glob("*.npz")
num_exps = len(filenames)
idx_exps = range(num_exps)

fracs = []
frac_obs = np.linspace(0.1, 0.9, 9)
for i in range(len(frac_obs)):
    frac_pop1 = np.arange(0.1, frac_obs[i]-0.09, 0.1)
    for j in range(len(frac_pop1)):
        fracs.append((frac_pop1[j], frac_obs[i], 1))

pi_method = 'heuristic'
num_iter = 300

for idx in [0,2,5,9]:
    
    print('idx', idx)
    file = 'LDS_save_idx' + str(idx) + '.npz'
    print(file)
    
    for prot in [0, 12]: #range(len(fracs)):
        loadfile = np.load(file)

        data = loadfile['y']
        T,p = data.shape
        stateseq = loadfile['x']
        n   = stateseq.shape[1]
        obs_scheme = {'sub_pops' : tuple([loadfile['sub_pops'][i] for i in range(3)]), 
                      'obs_pops' : loadfile['obs_pops'], 
                      'obs_time' : tuple([np.int(item * T) for item in fracs[prot]])}
        
        print('prot', prot)
        print('fracs obs time', fracs[prot])

        print('obs_time: ', obs_scheme['obs_time'])
        seq, seq_f = obs_scheme.copy(), {'sub_pops' : (np.arange(p),), 'obs_pops' : (0,), 'obs_time' : (T,)} 
        seq_f['y'], seq['y'] = data, data.copy()
        

        params, SIGfp, SIGyy, lPSIGfp, lPSIGyy, SIGfp_f, SIGyy_f = \
            stitching_ssid.run_exp_iterSSID(seq, seq_f, n, num_iter=num_iter,pi_method=pi_method, plot_flag=True)

        print('len(params):', len(params))
        params_naive = stitching_ssid.stabilize_A(params[0])
        params_final = stitching_ssid.stabilize_A(params[-1])

        save_file = '../init/init_prot' + str(prot) + '_' + file[:-4] 
        save_file_m = {'params': params_final,
                       'params_naive': params_naive,
                       'params_hist' : params,
                       'SIGfp' : SIGfp,
                       'SIGyy' : SIGyy,
                       'lPSIGfp' : lPSIGfp,
                       'lPSIGyy' : lPSIGyy,
                       'data': seq['y'],                   
                       'p' : p,
                       'n' : n,
                       'T' : T,
                       'filename' : file,
                       'num_iter' : num_iter,
                       'pi_method' : pi_method, 
                       'sub_pops': obs_scheme['sub_pops'],
                       'obs_time': obs_scheme['obs_time'],
                       'obs_pops': obs_scheme['obs_pops'],
                       'fracs': fracs,
                       'prot': fracs[prot]
                       }

        savemat(save_file,save_file_m) # does the actual saving

        np.savez(save_file, params=params_final,
                            params_naive=params_naive,
                            params_hist=params, 
                            SIGfp = SIGfp,
                            SIGyy = SIGyy,
                            lPSIGfp = lPSIGfp,
                            lPSIGyy = lPSIGyy,
                            data = seq['y'],                   
                            p = p,
                            n = n,
                            T = T,
                            filename = file,
                            num_iter = num_iter,
                            pi_method = pi_method,
                            sub_pops=obs_scheme['sub_pops'],            
                            obs_time=obs_scheme['obs_time'],            
                            obs_pops=obs_scheme['obs_pops'],
                            fracs=fracs, 
                            prot=fracs[prot])               
```

# Generate SSID initialisations for experiment #1

```python
%matplotlib inline
import numpy as np
from scipy.io import loadmat
import glob, os

absolute_code_path = '/home/marcel/Desktop/Projects/Stitching/code/code_le_stitch/iterSSID/python/'
os.chdir(absolute_code_path +'core')
import stitching_ssid

from scipy.io import savemat # store results for comparison with Matlab code   

relative_data_path = '../../../../../results/cosyne_poster/experiment_1/init'

print('loading data')
spikes = loadmat('/home/marcel/Desktop/Projects/Stitching/results/cosyne_poster/gb_net/spikes_20trials_10msBins')
spikes= spikes['spikes_out']

neuron_shuffle = np.random.choice(p,p,replace=False)

num_iter = 0
pi_method = 'heuristic'

print('concatenating trials')
p, T = 1000, 600
#data = spikes[0][0].T

print('number trials: ', spikes.size)
idx_n = np.sort(np.random.choice(1000, size=p, replace=False))
data = np.vstack([spikes[i][0].T[:,idx_n] for i in range(spikes.size)]).astype(np.float)
T *= spikes.size


### SHUFFLE NEURONS (ONLY ONCE!!)

data =  data[:,neuron_shuffle]
print('data.shape', data.shape)
###



n = 10

sub_pops = (np.arange(p//2+100), np.arange(p//2-100, p))
obs_pops = np.array((0,1))
obs_time = np.array((T//2,T))

obs_scheme = {'sub_pops' : sub_pops, 
              'obs_pops' : obs_pops, 
              'obs_time' : obs_time}

print('creating seq dictionaries')

seq, seq_f = obs_scheme.copy(), {'sub_pops' : (np.arange(p),), 'obs_pops' : (0,), 'obs_time' : (T,)} 
seq_f['y'], seq['y'] = data, data.copy()

print('starting fit')
params, SIGfp, SIGyy, lPSIGfp, lPSIGyy, SIGfp_f, SIGyy_f = \
    stitching_ssid.run_exp_iterSSID(seq, seq_f, n, num_iter=num_iter,pi_method=pi_method, plot_flag=True)

params_naive = stitching_ssid.stabilize_A(params[0])
params_final = stitching_ssid.stabilize_A(params[-1])
        
save_file = '/home/marcel/Desktop/Projects/Stitching/results/cosyne_poster/experiment_1/init/init_experiment' 
save_file_m = {'params': params_final,
               'params_naive': params_naive,
               'params_hist' : params,
               'SIGfp' : SIGfp,
               'SIGyy' : SIGyy,
               'lPSIGfp' : lPSIGfp,
               'lPSIGyy' : lPSIGyy,
               'data': seq['y'],                   
               'p' : p,
               'n' : n,
               'T' : T,
               'filename' : save_file,
               'num_iter' : num_iter,
               'pi_method' : pi_method,
               'neuron_shuffle': neuron_shuffle
               }

savemat(save_file,save_file_m) # does the actual saving

np.savez(save_file, params=params_final,
                    params_naive=params_naive,
                    params_hist=params, 
                    SIGfp = SIGfp,
                    SIGyy = SIGyy,
                    lPSIGfp = lPSIGfp,
                    lPSIGyy = lPSIGyy,
                    data = seq['y'],                   
                    p = p,
                    n = n,
                    T = T,
                    filename = save_file,
                    num_iter = num_iter,
                    pi_method = pi_method,
                    neuron_shuffle = neuron_shuffle,
                    sub_pops=obs_scheme['sub_pops'],            
                    obs_time=obs_scheme['obs_time'],            
                    obs_pops=obs_scheme['obs_pops'])  
```
